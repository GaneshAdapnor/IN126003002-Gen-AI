# **Task 1: Conceptual Understanding**

1)  **What is the difference between "Love" and "love" in NLP?**

**Ans :**
- Depends on use case and whether model should be case sensitive or insensitive meaning of the both tokens varies.

- case sensitive : Both words as two different tokens, meaning alters. Example : Named Entity Recognition, Sentiment nuance, noun detection

- case insensitive : Both tokens are treated as same, providing same meaning, vocabulary size reduces. Example : Spam detection, sentiment classification



2) **What happens if stopwords are not removed?**

- Stopwords are common words and Removing stopwords is optional

- If not removed, vocabulary size increases, training time will be huge, memory usage is high, noise in features, less important to highly significant word, sometimes reduce model performance due to complex computations

- If kept based on use case , preserves semantic meaning in advanced architectures like transformers, especially important in sentiment analysis




3) **Provide two real-world scenarios where removing stopwords can be harmful.**

- i) Sentiment Analysis : Negation handling - if stop words like no, never, not are removed, gives rise to wrong sentiment prediction in few tasks like reviews, feedback etc

- ii) Question Answering Chatbots/ Language translation : Removing stopwords might lose structure, intent, model performance reduced




4) **Explain the difference between stemming and lemmatization.**

- stemming : process of removing suffixes/prefixes from the word and making it closer to base form , may not give meaningful base word eg : Porter Stemmer

- Lemmatization : process of converting given word to base word that forms meaningful word using vocabulary, PoS tagging eg : WordNet Lemmatizer

- Stemming is faster than Lemmatization

```python
def preprocess_text(text):
  # Remove numbers
  text=re.sub(r'\d+','',text)

  #Remove extra spaces
  text=re.sub(r'\s+',' ',text).strip()

  #Handle repeated characters
  text=re.sub(r'(.)\1{2,}',r'\1',text)

  #Convert text to lowercase
  text=text.lower()

  #Remove URLs and email-like patterns
  text=re.sub(r'http\S+|www\S+','',text)
  text=re.sub(r'\S+@\S+','',text)

  #Remove special characters (keep only alphabets and space)
  text=re.sub(r'[^a-zA-Z\s]','',text)

  #Remove very short tokens
  tokens=text.split()
  tokens=[word for word in tokens if len(word)>2 or word in ['no','not']]



  return tokens,' '.join(tokens)


```

# **Task 7: Error Handling**


```python
def preprocess_text(text):

  if not isinstance(text,str): #Handling None or non string input
    return [],""

  if text.strip()=="":  #Handling empty string
    return [],""

  # Remove numbers
  text=re.sub(r'\d+','',text)

  #Remove extra spaces
  text=re.sub(r'\s+',' ',text).strip()

  #Handle repeated characters
  text=re.sub(r'(.)\1{2,}',r'\1',text)

  #Convert text to lowercase
  text=text.lower()

  #Remove URLs and email-like patterns
  text=re.sub(r'http\S+|www\S+','',text)
  text=re.sub(r'\S+@\S+','',text)

  #Remove special characters (keep only alphabets and space)
  text=re.sub(r'[^a-zA-Z\s]','',text)

  #if nothing left after cleaning
  if text=="":
    return [],""

  #Remove very short tokens
  tokens=text.split()
  tokens=[word for word in tokens if len(word)>2 or word in ['no','not']]



  return tokens,' '.join(tokens)
```

```python
def full_pipeline(text_list):
  all_tokens=[]
  clean_sentences=[]

  for text in text_list:
    tokens,cleaned=preprocess_text(text)
    all_tokens.extend(tokens)
    clean_sentences.append(cleaned)

  return {
      'tokens':all_tokens,
      'clean sentences':clean_sentences
  }
```
